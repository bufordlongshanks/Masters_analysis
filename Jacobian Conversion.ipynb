```python
df = pd.read_csv(os.listdir()[0], delim_whitespace=True, comment='#')
plt.plot(df.iloc[:,0],df.iloc[:,1])
```

```python
def to_energy(x,y):
    w = np.array(x)
    hc = 6.62e-34* 3e8
    e = hc/(w*1e-9)
    n = np.array(y)
    n_e = (hc/e**2)*n
    el = 1.60e-19
    e = e/el
    n_e = n_e/max(n_e)
    return e,n_e
    
    
```

```python
energy, counts = to_energy(df.iloc[:,0], df.iloc[:,1]-300)
plt.plot(energy,counts)
plt.xlim(2.0,2.8)
```
